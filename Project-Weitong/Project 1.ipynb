## Clean First Dataset as Netflix Movie from 2011 to 2021


```python
import os
import csv
import pandas as pd

#Read First csv file path: the netflix_titles
netflix_df = pd.read_csv("Resources/netflix_titles.csv")
netflix_df.head()
```

```python
#Take only the movie from the type column, we don't need TV show for our project
netflix_moive_df=netflix_df[netflix_df["type"] == "Movie"]
netflix_moive_df = netflix_moive_df.reset_index()
netflix_moive_df.head()
```

```python
#Take the columns that are useful for our project question
clean_moive_df = netflix_moive_df[["show_id","title", "director", "country", "release_year", "rating", "duration","listed_in","type"]]

#Rename the columns
clean_moive_df = clean_moive_df.rename(columns = {"show_id" : "Show_ID",
                                                  "title": "Moive Title",
                                                  "director": "Director",
                                                  "country": "Country",
                                                  "release_year":"Release_Year",
                                                  "rating" : "Rating_class",
                                                  "duration" : "Duration",
                                                  "listed_in" : "Categories",
                                                  "type" : "Type",})
print(clean_moive_df.index)
print(clean_moive_df.columns)

clean_moive_df.head(10)
```

```python
#Our project only need data period 2011-2021.
year_netflix_df= clean_moive_df[clean_moive_df["Release_Year"].isin([2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021])]
final_netflix_df=year_netflix_df.reset_index()
del final_netflix_df["index"]
final_netflix_df.head()
```

```python
#Check the type of each column to make sure if "Release_Year" and "Duration" as number
#There are 'NaN' in column "Moive Title", "Director" and "Duration"

print(final_netflix_df.dtypes)
```

```python
#Transfer the "Duration" from string to integer, if the column = 'NaN', then consider it as integer 0
import numpy as np
final_netflix_df['Duration'] =final_netflix_df['Duration'].str.split().str[0]
final_netflix_df['Duration'] = final_netflix_df['Duration'].apply(lambda x: int(x) if not np.isnan(float(x)) else 0)
print(final_netflix_df.dtypes)
```

```python
# Final data frame for Netflix Moive
final_netflix_df
```

```python
#Save the cleaned data for future analysis
final_netflix_df.to_csv('Final_Netflix_Movie.csv', index=False)
```

# 2.Clean the Second Dataset -- Netflix revenues quarterly from 2018-2020

```python
import pandas as pd

# read the original data
df = pd.read_csv("Resources/DataNetflixRevenue2020_V2.csv")

# create a pivot table
pivot_table = pd.pivot_table(df, values='Revenue', index=['Years'], columns=['Area'], aggfunc=sum)

# add a Grand Total column
pivot_table['Grand Total'] = pivot_table.sum(axis=1)

# add a Row Labels index
pivot_table.reset_index(inplace=True)

# display the pivot table
print(pivot_table.to_string(index=False))
```

```python
#Transfer pivot_table to DataFrame
revenue_df = pd.DataFrame(pivot_table)
revenue_df
```

```python
#split column 'Years' to 'Quarter', 'Year'
revenue_df[['Quarter', 'Year']] = revenue_df['Years'].str.split(' - ', expand=True)
del revenue_df["Years"]
```

```python
#Check the types of all columns
print(revenue_df.dtypes)
```

```python
#Change'Year' column from string to integer 
revenue_df['Year'] = revenue_df['Year'].astype(int)
print(revenue_df.dtypes)
```

```python
#Final revenue table
revenue_df
```

```python
#Save the file as a new csv for future analysis
revenue_df.to_csv('Final_Revenues.csv', index=True)
```
