## Clean First Dataset as Netflix Movie from 2011 to 2021


```python
import os
import csv
import pandas as pd

#Read First csv file path: the netflix_titles
netflix_df = pd.read_csv("Resources/netflix_titles.csv")
netflix_df.head()
```

```python
#Take only the movie from the type column, we don't need TV show for our project
netflix_movie_df=netflix_df[netflix_df["type"] == "Movie"]
netflix_movie_df = netflix_movie_df.reset_index()
netflix_movie_df.head()
```

```python
#Take the columns that are useful for our project question
clean_movie_df = netflix_movie_df[["show_id","title", "director", "country", "release_year", "rating", "duration","listed_in","type"]]

#Rename the columns
clean_movie_df = clean_movie_df.rename(columns = {"show_id" : "Show_ID",
                                                  "title": "Movie Title",
                                                  "director": "Director",
                                                  "country": "Country",
                                                  "release_year":"Release_Year",
                                                  "rating" : "Rating_class",
                                                  "duration" : "Duration",
                                                  "listed_in" : "Categories",
                                                  "type" : "Type",})
print(clean_movie_df.index)
print(clean_movie_df.columns)

clean_movie_df.head(10)
```

```python
#Our project only need data period 2011-2021.
year_netflix_df= clean_movie_df[clean_movie_df["Release_Year"].isin([2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021])]
final_netflix_df=year_netflix_df.reset_index()
del final_netflix_df["index"]
final_netflix_df.head()
```

```python
#Check the type of each column to make sure if "Release_Year" and "Duration" as number
#There are 'NaN' in column "Moive Title", "Director" and "Duration"

print(final_netflix_df.dtypes)
```

```python
#Transfer the "Duration" from string to integer, if the column = 'NaN', then consider it as integer 0
import numpy as np
final_netflix_df["Duration"] =final_netflix_df["Duration"].str.split().str[0]
final_netflix_df["Duration"] = final_netflix_df["Duration"].apply(lambda x: int(x) if not np.isnan(float(x)) else 0)
print(final_netflix_df.dtypes)
```

```python
# Final data frame for Netflix Moive
final_netflix_df
```

```python
#Save the cleaned data for future analysis
final_netflix_df.to_csv('Clean_Netflix_Movie.csv', index=False)
```

## 2. Clean the Second Dataset and merge with First Dataset

```python
#Read First csv file path
rating_df = pd.read_csv("Resources/titles.csv")
rating_df.head()
```

```python
netflix_rating_df=rating_df[rating_df["type"] == "MOVIE"]
netflix_rating_df = netflix_rating_df.reset_index()
netflix_rating_df.head(5)
```

```python
year_rating_df= netflix_rating_df[netflix_rating_df["release_year"].isin([2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021])]
year_rating_df=year_rating_df.reset_index()
del year_rating_df["index"]
year_rating_df.head()
```

```python
clean_rating_df = year_rating_df[["title", "imdb_score", "imdb_votes"]]
clean_rating_df = clean_rating_df.rename(columns = {"title" : "Movie Title",
                                                   "imdb_score" : "IMDB_Score",
                                                   "imdb_votes" : "IMDB_Votes"})

clean_rating_df.head()
```

```python
#Check the type of each column to 
#If there are 'NaN' in column "IMDB_Score"and "IMDB_Votes", then consider as 0

print(clean_rating_df.dtypes)
```

```python
#Save the clean data frame as 'Netflix_Rating.csv'
clean_rating_df.to_csv("Netflix_Rating.csv", index=False)
```

## Merge the two data set and save as 'Merged_Netflix_movie&rating.csv'

```python

title_rating_df = pd.merge(final_netflix_df, clean_rating_df, on = ["Movie Title"] , how = "inner")
title_rating_df
```

```python
#save the merged Data Frame as 'Merged_Netflix_movie&rating.csv' for analysis
title_rating_df.to_csv("Merged_Netflix_movie&rating.csv", index=False)
```
