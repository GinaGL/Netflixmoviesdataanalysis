```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# File to Load (Remember to Change These)
movie_netflix = "Merged_Netflix_movie&rating.csv"
stock_netflix = "Final Merged Dataset.csv"
```

```python
# Read movie data Files and store into Pandas DataFrames
m_netflix = pd.read_csv(movie_netflix)
s_netflix = pd.read_csv(stock_netflix)

s_netflix
```

```python

s_netflix2 = s_netflix.set_index("Years")

NA_netflix = s_netflix2[(s_netflix2["Area"] == "United States and Canada")]

Asia_netflix = s_netflix2[(s_netflix2["Area"] == "Asia-Pacific")]

EurAF_netflix = s_netflix2[(s_netflix2["Area"] == "Europe,  Middle East and Africa")]

Latin_netflix = s_netflix2[(s_netflix2["Area"] == "Latin America")]

Latin_netflix

```

```python

 
plt.plot( Latin_netflix['Revenue'], color='blue')
plt.plot( NA_netflix['Revenue'], color='red')
plt.plot( EurAF_netflix['Revenue'], color='green')
plt.plot( Asia_netflix['Revenue'], color='orange')

plt.xticks( rotation='60')

plt.legend(["Latin America", "North America" , "Europe", 'Asia'])

plt.tight_layout()

plt.title('Netflix Revenue by Region, 2018-2022', fontsize=12)
plt.xlabel('Quarters', fontsize=12)
plt.ylabel('Revenue', fontsize=12)

plt.show()

```

```python


labels = NA_netflix.index

data1 = NA_netflix["Close"]

width = 0.8


plt.xlabel("Quarters")
plt.ylabel("Closing Price of Netflix Stock ($)")
plt.title("Closing Price of Netflix Stock, 2018-2022")

plt.xticks(range(len(data1)), labels, rotation = "50")

plt.bar(np.arange(len(data1)), data1, width=width) 


plt.tight_layout()



plt.show()
```

```python
plt.plot( Latin_netflix['Subscribers'], color='blue')
plt.plot( NA_netflix['Subscribers'], color='red')
plt.plot( EurAF_netflix['Subscribers'], color='green')
plt.plot( Asia_netflix['Subscribers'], color='orange')

plt.xticks( rotation='60')

plt.legend(["Latin America", "North America" , "Europe", 'Asia'])

plt.tight_layout()

plt.title('Number of Netflix Subscribers by Region, 2018-2022', fontsize=12)
plt.xlabel('Quarters', fontsize=12)
plt.ylabel('Number of Subscribers (10,000,000)', fontsize=12)

plt.show()
```
