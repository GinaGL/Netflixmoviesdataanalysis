```python
import pandas as pd
```

```python
#read csv files and merge revenue df and subscriber df
df1 = pd.read_csv("Netflix Revenue.csv")
df2 = pd.read_csv("Netflix Subscriber.csv")
df3 = pd.read_csv("NetflixHistoricStockPrice.csv")
df4 = pd.merge(df1, df2, on=["Area", "Years"]).drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y'])
df4["Area"].unique()
```

```python
df4_UnitedStateandCanada = df4[df4["Area"] == "United States and Canada"].reset_index(drop=True)
df4_EuropeMiddleEastandAfrica = df4[df4["Area"] == "Europe,  Middle East and Africa"].reset_index(drop=True)
df4_LatinAmerica = df4[df4["Area"] == "Latin America"].reset_index(drop=True)
df4_AsiaPacific = df4[df4["Area"] == "Asia-Pacific"].reset_index(drop=True)
df4_EuropeMiddleEastandAfrica
```

```python
#filter only the month at quarter end
df3["Date"] = pd.to_datetime(df3["Date"])
df3_quarterly = df3[df3["Date"].dt.strftime('%m-%d').isin(['04-01', '01-01', '07-01', '10-01'])]
df3_quarterly
```

```python
#remove first row as it's the 2017 Q4 data and remove Date column
df3_quarterly_final = df3_quarterly[df3_quarterly["Date"] != "2018-01-01"].reset_index(drop=True)
df3_quarterly_final
```

```python
#merge quarterly stock date with merged revenue/subscriber df
df4_UnitedStateandCanada_merged = pd.merge(df4_UnitedStateandCanada,df3_quarterly_final, left_index=True, right_index=True)
df4_EuropeMiddleEastandAfrica_merged = pd.merge(df4_EuropeMiddleEastandAfrica, df3_quarterly_final, left_index=True, right_index=True)
df4_LatinAmerica_merged = pd.merge(df4_LatinAmerica,df3_quarterly_final, left_index=True, right_index=True)
df4_AsiaPacific_merged = pd.merge(df4_AsiaPacific,df3_quarterly_final, left_index=True, right_index=True)
df4_UnitedStateandCanada_merged
```

```python
#merge all areas together and delete Date column
df_cleaned = pd.concat([df4_UnitedStateandCanada_merged, df4_EuropeMiddleEastandAfrica_merged, df4_LatinAmerica_merged, df4_AsiaPacific_merged])
del df_cleaned["Date"]
df_cleaned= df_cleaned.reset_index(drop=True)
df_cleaned
```

```python
#export to csv
df_cleaned.to_csv("Final Merged Dataset.csv")
```
