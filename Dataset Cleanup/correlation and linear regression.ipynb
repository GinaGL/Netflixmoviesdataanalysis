```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
df_cleaned = pd.read_csv("Final Merged Dataset.csv")
df_cleaned.head()
```

```python
#Generate a scatter plot of average revenue vs. stock price

average_revenue = df_cleaned.groupby("Years")["Revenue"].mean()
stock_price = df_cleaned.groupby("Years")["Close"].mean()

plt.scatter(average_revenue, stock_price, color='blue')
plt.xlabel("Revenue")
plt.ylabel("Stock Price")
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model 
# for average revenue vs.stock price
corr = average_revenue.corr(stock_price).round(2)
print(f"The correlation between Revenue and Stock Price is {corr}")

pe_slope, pe_int, pe_r, pe_p, pe_std_err = st.linregress(average_revenue, stock_price)
pe_fit = pe_slope * stock_price + pe_int

plt.figure(figsize=(6,4))
plt.scatter(average_revenue, stock_price, color='blue')
plt.plot(average_revenue,pe_fit, color = "red")
plt.xlabel('Revenue')
plt.ylabel('Stock Price')
plt.show()
```
