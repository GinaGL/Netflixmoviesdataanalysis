```python
#  Generate a bar plot showing the total amount of movie in each categoery 

data = {'Categories': ['Documentaries', 
                       'Children & Family Movies', 
                       'Comedies, Dramas', 
                       'Dramas, International Movies', 
                       'Children & Family Movies, Comedies'],
        'Count': [1000, 2000, 3000 , 4000, 1000]}

df = pd.DataFrame(data)

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

plt.bar(df['Categories'], df['Count'], color=colors)
plt.xticks(rotation=45)
plt.xlabel('Categories')
plt.ylabel('Count (Hundreds)')
plt.title('Number of Movies by Category')
plt.show()
```

```python
data = {'Categories': ['Documentaries', 
                       'Children & Family Movies', 
                       'Comedies, Dramas', 
                       'Dramas, International Movies', 
                       'Children & Family Movies, Comedies'],
        'Count': [1000, 2000, 3000 , 4000, 1000]}

df = pd.DataFrame(data)

plt.pie(df['Count'], labels=df['Categories'], autopct='%1.1f%%')
plt.title('Distribution of Content Categories')
plt.show()
```

```python
data = {'Rating_class': ['PG-13', 'PG', 'PG-13', 'TV-MA', 'TV-PG']}
df = pd.DataFrame(data)

value_counts = df['Rating_class'].value_counts()

plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Content Rating')
plt.legend(title='Rating Class', loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
#  Generate a pie showing the distribution of IMDB Scores

movie_rating_df = pd.read_csv('Project-Weitong/Merged_Netflix_movie&rating.csv')

bins = [0, 5, 8, 10]
labels = ['<5', '5-8', '>8']

score_ranges = pd.cut(movie_rating_df['IMDB_Score'], bins=bins, labels=labels, include_lowest=True)
score_counts = score_ranges.value_counts()

plt.pie(score_counts, labels=score_counts.index, autopct='%1.1f%%')
plt.title('Distribution of IMDB Scores')
plt.legend(title='IMDB Score', loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
